```python
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
```

# Reading Corpus

```python
corpus_file = open('corpus.txt', mode='r')
corpus = corpus_file.read()
print(corpus[:45])
```

```python
vocab = set(corpus)
num_classes = len(vocab)

vocab_to_int = {c:i for i, c in enumerate(vocab)}
int_to_vocab = {i:c for i, c in enumerate(vocab)}

encoded = [vocab_to_int[c] for c in corpus]

print(encoded[:45])
print(len(encoded))
```

```python
# Hyperparams
batches = 1000
steps = 50
dropout = 0.5
epochs = 1000
layers = [steps, 50, 100, steps]

X = np.array(encoded[:batches*steps])
y = np.array(encoded[1:batches*steps + 1])

X_train = X.reshape(batches, steps)
y_train = y.reshape(batches, steps)

X_one_hot = np.eye(num_classes)[X_train]
y_one_hot = np.eye(num_classes)[y_train]

print(X_train.shape)
print(X_one_hot.shape)
print(y_train.shape)
print(y_one_hot.shape)
```

```python
model = Sequential()
model.add(LSTM(100,input_shape=(None, 76),return_sequences=True))
model.add(Dropout(dropout))
model.add(LSTM(200,return_sequences=True))
model.add(Dropout(dropout))
model.add(Dense(76))
model.add(Activation("softmax"))
model.compile(loss="categorical_crossentropy", optimizer="adam")
```

```python
model.fit(X_one_hot, y_one_hot, batch_size=100, epochs=epochs)
```

```python
y_pred = model.predict(np.reshape(X_one_hot[1,:,:],(1,50,76)))
y_transformed = np.argmax(y_pred.reshape(25,76), axis=1)
print(y_transformed)
```

```python
output = [int_to_vocab[i] for i in y_transformed]
print(''.join(output))
print(corpus[0:26])
```
